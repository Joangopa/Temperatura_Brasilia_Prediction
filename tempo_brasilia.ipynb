### Modelo Sequential

```python
# [[[1], [2], [3], [4], [5]]] [6]
# [[[2], [3], [4], [5], [6]]] [7]
# [[[3], [4], [5], [6], [7]]] [8]

def df_to_X_y(df, window_size=5):
  df_as_np = df.to_numpy()
  X = []
  y = []
  for i in range(len(df_as_np)-window_size):
    row = [[a] for a in df_as_np[i:i+window_size]]
    X.append(row)
    label = df_as_np[i+window_size]
    y.append(label)
  return np.array(X), np.array(y)
```

```python
WINDOW_SIZE = 5
X1, y1 = df_to_X_y(temp_max_combined, WINDOW_SIZE)
X1.shape, y1.shape
```

```python
X_train1, y_train1 = X1[:30000], y1[:30000]
X_val1, y_val1 = X1[30000:35000], y1[30000:35000]
X_test1, y_test1 = X1[35000:], y1[35000:]
X_train1.shape, y_train1.shape, X_val1.shape, y_val1.shape, X_test1.shape, y_test1.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam

model1 = Sequential()
model1.add(InputLayer((5, 1))) # type: ignore
model1.add(LSTM(64)) # type: ignore
model1.add(Dense(8, 'relu')) # type: ignore
model1.add(Dense(1, 'linear')) # type: ignore

model1.summary()
```

```python
model1.save("model1.keras")
```

```python
cp1 = ModelCheckpoint('model1.keras', save_best_only=True)
```

```python
model1.compile(loss=MeanSquaredError(), optimizer=Adam(learning_rate=0.0001), metrics=[RootMeanSquaredError()])
```

```python
model1.fit(X_train1, y_train1, validation_data=(X_val1, y_val1), epochs=10, callbacks=[cp1])
```

```python
train_predictions = model1.predict(X_train1).flatten()
train_results = pd.DataFrame(data={'Train Predictions':train_predictions, 'Actuals':y_train1})
train_results
```

```python

plt.plot(train_results['Train Predictions'][50:100], label = 'Train Predictions')
plt.plot(train_results['Actuals'][50:100], label = 'Actuals')
plt.legend(loc='upper right')
```

```python
import numpy as np
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

# Suponiendo que 'series' es tu serie temporal original y 'series_diff' es la serie diferenciada
# Identificar p y q
plot_pacf(temp_max_combined)
plt.show()

# Ajustar varios modelos
best_aic = np.inf
best_order = None
best_model = None

for p in range(7):  # Prueba valores de 0 a 4 para p
    for q in range(7):  # Prueba valores de 0 a 4 para q
        try:
            model = ARIMA(temp_max_combined, order=(p, 1, q))  # Usando d=1 según diferenciación
            model_fit = model.fit()
            print(f'ARIMA({p}, 1, {q}) - AIC:{model_fit.aic}')
            if model_fit.aic < best_aic:
                best_aic = model_fit.aic
                best_order = (p, 1, q)
                best_model = model_fit
        except:
            continue

print(f'Mejor Modelo: ARIMA{best_order} - AIC:{best_aic}')
print(best_model.summary())

# Diagnóstico de residuos del mejor modelo
residuals = best_model.resid
plt.figure()
plt.subplot(211)
plot_acf(residuals, ax=plt.gca())
plt.subplot(212)
plot_pacf(residuals, ax=plt.gca())
plt.show()
```
